# Head stabilization inference

`flygym_demo.examples.head_stabilization` ships **`JointAngleScaler`**, **`ThreeLayerMLP`**, and **`HeadStabilizationInferenceWrapper`**: given proprioception + leg contact flags, the network predicts compensatory head motion (v2 tutorial port of the 1.x demo).

Weights live under `flygym_demo.examples.assets.trained_models.head_stabilization/`. Loading the checkpoint requires **PyTorch** (optional dev dependency). Without it, `HeadStabilizationInferenceWrapper()` still runs and returns zeros (see unit tests).


```python
import numpy as np
from flygym_demo.examples.head_stabilization import HeadStabilizationInferenceWrapper

wrapper = HeadStabilizationInferenceWrapper()
ja = np.zeros(42, dtype=np.float32)
cf = np.zeros(6, dtype=np.float32)
print(wrapper.predict(ja, cf))
```

With `torch` installed, load the scaler and MLP state dict from `all_dofs_model.ckpt` and pass them into `HeadStabilizationInferenceWrapper(scaler=..., model=...)`. Match `input_dim` to the length of `JointAngleScaler.mean`.
