# Advanced olfaction: multiple cues and plume movies

For **multiple odor channels** at distinct sources, pass a `(n_sources, n_channels)` `peak_odor_intensity` matrix into **`OdorWorld`** (same API as tutorial 9).

For **time-varying plumes** stored as HDF5 movies (`plume` dataset), use **`OdorPlumeWorld`** in `flygym_demo.examples.worlds.sensory`. Point `plume_data_path` at your file and install `h5py`. The FlyGym 1.x tutorials generated such datasets with separate scripts; here we only show the world interface alongside the static multi-source example.


```python
import numpy as np
from flygym import Simulation
from flygym.anatomy import ContactBodiesPreset
from flygym.utils.math import Rotation3D
from flygym_demo.examples import make_walking_fly, run_closed_loop
from flygym_demo.examples.locomotion import HybridTurningController
from flygym_demo.examples.olfaction import SimpleOdorTaxisController, add_odor_sensors
from flygym_demo.examples.worlds import OdorWorld

fly = make_walking_fly(add_camera=False)
sites = add_odor_sensors(fly)
world = OdorWorld(
    odor_source=np.array([[12.0, -1.5, 0.0], [12.0, 1.5, 0.0]]),
    peak_odor_intensity=np.array([[1.0, 0.2], [0.2, 0.9]]),
)
world.add_fly(
    fly,
    spawn_position=[0, 0, 1.5],
    spawn_rotation=Rotation3D("quat", [1, 0, 0, 0]),
    bodysegs_with_ground_contact=ContactBodiesPreset.TIBIA_TARSUS_ONLY,
    add_ground_contact_sensors=False,
)
sim = Simulation(world)
base = HybridTurningController(sim.timestep)
controller = SimpleOdorTaxisController(base, sites, attractive_gain=0.5, aversive_gain=0.5)
records = run_closed_loop(sim, controller, 0.005, fly_name=fly.name)
print("odor shape", np.asarray(records[-1]["metadata"]["odor"]).shape)
```

Optional: `from flygym_demo.examples.worlds.sensory import OdorPlumeWorld`.
