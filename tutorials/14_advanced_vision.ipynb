# Advanced vision: compound-eye preprocessing

This notebook highlights **extra Retina utilities** used in vision pipelines: **`correct_fisheye`** and the hexagonal **`raw_image_to_hex_pxls` / `hex_pxls_to_human_readable`** mapping (see Lam et al. and the FlyGym 1.x vision tutorials for biological motivation).

For closed-loop object following with **`SimpleVisualTaxisController`**, see [tutorial 11](11_vision_basics.ipynb).


```python
import numpy as np
from flygym_demo.examples.vision import Retina

retina = Retina()
img = np.zeros((retina.nrows, retina.ncols, 3), dtype=np.uint8)
img[:, : retina.ncols // 2, 1] = 200
img[:, retina.ncols // 2 :, 2] = 180
undist = retina.correct_fisheye(img)
left = retina.raw_image_to_hex_pxls(undist[:, : undist.shape[1] // 2])
right = retina.raw_image_to_hex_pxls(undist[:, undist.shape[1] // 2 :])
print(left.shape, right.shape, left.max(), right.max())
```
